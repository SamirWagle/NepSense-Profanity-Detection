```python
import pandas as pd
facebook_datas=pd.read_csv("./Facebook Datas/dataset_facebook-comments-scraper_2024-05-04_06-05-34-173.csv")
facebook_datas
```

```python
time_data=facebook_datas["date"]
```

```python
import pandas as pd
from datetime import datetime

# Function to convert string to datetime and extract hour, minute, and seconds
def process_date(date_string):
    date_object = datetime.strptime(date_string, '%Y-%m-%dT%H:%M:%S.%fZ')
    hour = date_object.hour
    minute = date_object.minute
    seconds = date_object.second
    return hour, minute, seconds

# Apply the function to the specified column
facebook_datas[['hour', 'minute', 'seconds']] = facebook_datas['date'].apply(lambda x: pd.Series(process_date(x)))

# Display the modified DataFrame
facebook_datas
```

```python
# Count occurrences of each hour
hour_counts = facebook_datas['hour'].value_counts().sort_index()

# Print count of occurrences for each hour
print("Hour\tCount")
for hour, count in hour_counts.items():
    print(f"{hour}\t{count}")
```

```python
import matplotlib.pyplot as plt
plt.hist(facebook_datas['hour'], bins=24, color='skyblue', edgecolor='black')
plt.xlabel('Hour of the Day')
plt.ylabel('Frequency')
plt.title('Distribution of Hours')
plt.xticks(range(24))
plt.grid(True)
plt.show()
```

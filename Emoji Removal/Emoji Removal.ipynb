```python
import pandas as pd
import re

# Load the CSV file containing the list of emoji unicodes
emoji_df = pd.read_csv('Emoji Sheets - Emoji Only.csv')

# Extract the emojis into a list
emoji_list = emoji_df['Emoji_List'].tolist()
emoji_list
```

```python
import re

# Start the pattern string
pattern = '['

# Append each code point to the pattern string, ensuring each one is 8 digits
for cp in emoji_list:
    pattern += f'\\U{cp[1:]:0>8}'

# Close the pattern string
pattern += ']'

# Compile the regular expression
emoji_pattern = re.compile(pattern, re.UNICODE)

# Output the pattern string for verification
print(emoji_pattern.pattern)
```

```python

# Compile the regular expression
emoji_pattern = re.compile(pattern, re.UNICODE)

# Function to remove emojis from a text string
def remove_emojis(text):
    return emoji_pattern.sub(r'', text)

# Example text containing emojis
example_text = "Here is a text with emojis üòÄü•∞ *Ô∏è‚É£ 0Ô∏è‚É£ 1Ô∏è‚É£ 2Ô∏è‚É£ 3Ô∏è‚É£ 4Ô∏è‚É£ 5Ô∏è‚É£ 6Ô∏è‚É£ 7Ô∏è‚É£ 8Ô∏è‚É£ 9Ô∏è‚É£ ¬©"

# Remove emojis from the example text
cleaned_text = remove_emojis(example_text)

print("Original text:", example_text)
print("Cleaned text:", cleaned_text)
```

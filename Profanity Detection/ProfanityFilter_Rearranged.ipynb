### Conversion Function for conversion of English Text to Nepali

```python
from Preprocessing.Nepali_nlp.Nepali_nlp.unicode_nepali import Unicode
import pandas as pd
from langdetect import detect, LangDetectException
from nepali_unicode_converter.convert import Converter
from tqdm import tqdm

# Define your conversion functions
def nepali_nlp_text_conversion(text):
    if isinstance(text, str) and text.strip():
        try:
            lang = detect(text)
            if lang != "ne":
                return Unicode().unicode_word(text)
        except LangDetectException:
            pass
    return text

def ai4bharat_text_conversion(text):
    if isinstance(text, str) and text.strip():
        temp_results = e.translit_sentence(text)["ne"]
        return temp_results
    return text

# Define the function that applies conversions to the DataFrame
def nepali_nlp_conversion(df):
    converter = Converter()

    # Wrap the apply functions with tqdm for the progress bar
    tqdm.pandas(desc="Applying Nepali NLP Conversion")
    df["nepali_nlp_conversion"] = df["text"].progress_apply(nepali_nlp_text_conversion)

    tqdm.pandas(desc="Applying Unicode Conversion Conversion")
    df["unicode_conversion"] = df["text"].progress_apply(lambda x: converter.convert(x) if isinstance(x, str) else '')

    tqdm.pandas(desc="Applying ai4bharat Conversion")
    df["ai4bharat"] = df["text"].progress_apply(ai4bharat_text_conversion)
    return df
```

# Profanity and offensive filtering

### Profanity and offesive filtering function

```python
import re
import pandas as pd
from tqdm import tqdm

def profanity_offensive_filter(df, to_be_tested_column_name, output_column_name, matched_pattern_column_name, filterwords):

    # Create a regex pattern from the filterwords | acts as 'OR' for regex at the start of the string or after whitespace
    pattern =r'(^|\s)(?:' + '|'.join(map(re.escape, filterwords)) + r')'
    #  r'^(?:' + '|'.join(map(re.escape, filterwords)) + r')'+

    df[to_be_tested_column_name] = df[to_be_tested_column_name].fillna('')

    # Filter the DataFrame using the regex pattern and specified column
    # df['match'] = df[to_be_tested_column_name].str.extract(f'({pattern})', flags=re.IGNORECASE, expand=False)
     # Find all matches for the pattern
    df["match"] = df[to_be_tested_column_name].str.findall(f'({pattern})', flags=re.IGNORECASE).apply(
        lambda matches: ', '.join(match[0] if isinstance(match, tuple) else match for match in matches)
    )

    # Create an empty list to store the results
    results = []

    # Iterate over the rows of the DataFrame with matches using tqdm progress bar
    matched_df = df.dropna(subset=['match'])
    for index, row in tqdm(matched_df.iterrows(), total=matched_df.shape[0], desc="Processing rows"):
        match = row['match']
        # Check if the match is in filterwords
        if match and any(re.search(pattern, match, re.IGNORECASE) for word in filterwords):
            results.append({
                to_be_tested_column_name: row[to_be_tested_column_name],
                output_column_name: 1,
                matched_pattern_column_name: match
            })

    # Convert the results list to a DataFrame
    results_df = pd.DataFrame(results)

    # Add the two new columns to the original DataFrame with progress bar
    tqdm.pandas(desc="Detecting")

    df[matched_pattern_column_name] = df['match']
    df[output_column_name] = df[matched_pattern_column_name].apply(lambda x: 1 if x!="" else 0)
    # df[output_column_name] = df[matched_pattern_column_name].apply(lambda x: 1 if pd.notna(x) else 0)

    # Drop the temporary 'match' column
    df = df.drop(columns=['match'])

    return matched_df, results_df, df
```

# Gender Detection

### Importing files

```python
import pandas as pd

# Load boys' names and girls' names from Excel files
combined_names_df = pd.read_csv("./Profanity Detection Datas/Combined Gender Datas.csv")
combined_names_df
```

### Conversion of Nepali name into English

```python
# %pip install nepali-to-roman
# %pip install langdetect
from langdetect import detect
import re
import ntr

def capitalize_after_space(name):
    words = name.split()
    capitalized_words = [word.capitalize() for word in words]
    return ' '.join(capitalized_words)

def detect_nep_and_conversion(text):
      # Ensure that text is a string or convert it to a string if it's not
  if not isinstance(text, str):
      text = str(text)
  sentences = re.split(r'(?<=[.!?])\s+(?=\D)', text)
  english_comments = []
  for sentence in sentences:
      try:
          language = detect(sentence)
          if language == "ne":
            sentence=capitalize_after_space(ntr.nep_to_rom(sentence))
            english_comments.append(sentence)
          else:
            english_comments.append(sentence)
      except:
          pass
  filtered_comment = '.'.join(english_comments)
  return filtered_comment
```

```python
# Apply the function with a progress bar
tqdm.pandas()

filtered_df["English_Name"] = filtered_df["profileName"].progress_apply(detect_nep_and_conversion)
filtered_df.head()
```

### Detection of Gender using Name

```python
# Function to get gender based on name input
def get_gender_from_name(name):
    if ' ' in name:
        name = name.split(maxsplit=1)[0].capitalize()
    else:
        name = name.capitalize()
    gender = combined_names_df[combined_names_df['Name'] == name]['Gender'].values
    if len(gender) > 0:
        return gender[0]
    else:
        return 'Unknown'
```

```python
# Apply the function with a progress bar
tqdm.pandas()

filtered_df["Gender"]=filtered_df["English_Name"].progress_apply(get_gender_from_name)
filtered_df[["English_Name","Gender"]].head()
```

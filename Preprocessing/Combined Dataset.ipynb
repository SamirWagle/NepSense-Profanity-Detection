```python
import os

# Get the current working directory
manual_tagged_data_directory = "../Profanity and Gender filtered Datas/Manual Tagged Datas"

# List all files in the current directory
file_name_list = []

# Iterate over all items in the directory
for item in os.listdir(manual_tagged_data_directory):
    # Check if the item is a file
    if os.path.isfile(os.path.join(manual_tagged_data_directory, item)):
        file_name_list.append(item)

file_name_list
```

```python
import pandas as pd

# Initialize the global variable
combined_df_length = 0

# Function to read an Excel file into a DataFrame with error handling
def read_excel_file(file_name):
    global combined_df_length
    try:
        df = pd.read_excel(f"{manual_tagged_data_directory}/{file_name}")
        combined_df_length += len(df)
        print(f"{file_name}'s length: {len(df)}")
        return df
    except Exception as e:
        print(f"Error reading {file_name}: {e}")
        return pd.DataFrame()  # Return an empty DataFrame in case of error

# Read all Excel files into DataFrames and combine them horizontally
dataframes = [read_excel_file(file_name) for file_name in file_name_list]
```

```python
combined_manual_tagged_df = pd.concat(dataframes, axis=0)
print(f"Combined DataFrame length: {combined_df_length}")
```

```python
from datetime import datetime, timedelta

# Define the base path and file name
base_path = "./Preprocessing Datas/"
base_filename = "Combined Manual Tagged Data"
latest_file_path=f"{base_path}/Combined Manual Tagged Data Latest.xlsx"

# Get the current UTC time
current_utc_time = datetime.utcnow()

# Convert to Nepal Time (UTC+5:45)
nepal_time = current_utc_time + timedelta(hours=5, minutes=45)
formatted_nepal_time = nepal_time.strftime("%Y-%m-%dT%H:%M:%S")

# Combine the base path, filename, and formatted Nepal time to create the full file path
combined_manual_tagged_output_path = f"{base_path}{base_filename}_{formatted_nepal_time}.xlsx"

# Output the new file path
combined_manual_tagged_output_path,latest_file_path
```

```python
import os
os.makedirs(os.path.dirname(combined_manual_tagged_output_path), exist_ok=True)
combined_manual_tagged_df.to_excel(combined_manual_tagged_output_path)
combined_manual_tagged_df.to_excel(latest_file_path)
# Print a success message
print(f"File saved successfully (path: {combined_manual_tagged_output_path})")
print(f"File saved successfully (path: {latest_file_path})")
```
